```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df
#To avoid encoding error, use 'unicode_escape'
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis= 1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int') 
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'shaadi'})
```

```python
df[['Age','Orders','Amount']].describe()
```

# Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df,edgecolor='black')
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen,edgecolor='black')
```

#### From above graphs we can see that most of thr buyers are even the purchasing power of females are greater than men

# Age

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender',edgecolor='black')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount', data = sales_age,color = 'tab:orange',edgecolor='black')
```

#### From above graphs we can see that most of the buyers are of age group between 26-35 yrs Female

# state

```python
# Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(17.90,10)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders', color = 'tab:orange',edgecolor='black')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17.90,5)})
sns.barplot(data = sales_state, x = 'State',y = 'Amount',color = 'tab:orange',edgecolor='black')
```

#### From above graphs we can see that unexpectedly most of orders are form Utter Pradesh, Maharashtra, Karanataka respectively but total sales/amount is from UP, Karnataka and the Maharashtra

# Marital Sratus 

```python
ax = sns.countplot(data = df, x = 'Marital_Status',edgecolor='black')
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y = 'Amount',hue = 'Gender',edgecolor='black')
```

#### From above graphs we can see that most of the buyers are married (Women) and they have high purchasing power

# Occupation

```python
sns.set(rc={'figure.figsize':(20,10)})
ax = sns.countplot(data = df, x = 'Occupation',edgecolor='black',color = 'tab:orange')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(data = sales_status, x = 'Occupation',y = 'Amount',edgecolor='black',color = 'tab:orange')
sns.set(rc={'figure.figsize':(20,10)})
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aiation  sector

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',edgecolor='black',color = 'tab:orange')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',edgecolor='black',color = 'tab:orange')
```

#### From above graphs we can see the most of the sold products are from Food,Clothig and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',edgecolor='black',color = 'tab:orange')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar',edgecolor='black',color = 'tab:orange')
```
